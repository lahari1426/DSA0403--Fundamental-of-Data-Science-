```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create Dataset
data = {
    'Patient_ID': ['P01','P02','P03','P04','P05','P06','P07','P08','P09','P10',
                   'P11','P12','P13','P14','P15','P16','P17','P18','P19','P20'],
    'Age': [45,32,55,28,60,40,52,35,48,30,62,38,50,27,58,42,46,34,53,29],
    'Sugar_Level': [145,98,180,90,200,120,165,105,155,95,210,115,170,88,190,130,150,100,175,92],
    'BP': [130,118,145,110,150,125,140,120,135,115,155,122,142,108,148,128,132,116,144,112],
    'BMI': [28,23,31,22,34,26,30,24,29,23,35,25,30,21,33,27,28,24,31,22],
    'Disease_Status': ['Yes','No','Yes','No','Yes','No','Yes','No','Yes','No',
                       'Yes','No','Yes','No','Yes','No','Yes','No','Yes','No']
}

# Create DataFrame
df = pd.DataFrame(data)

# Display Dataset
print("Healthcare Analytics Dataset")
print(df)

# Dataset Information
print("\nDataset Information")
print(df.info())

# Summary Statistics
print("\nSummary Statistics")
print(df.describe())

# Missing Values
print("\nMissing Values")
print(df.isnull().sum())

# Disease Status Count
print("\nDisease Status Count")
print(df['Disease_Status'].value_counts())

# -----------------------------
# Visualization
# -----------------------------

# 1. Age Distribution
plt.figure(figsize=(6,4))
plt.hist(df['Age'], bins=6)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.show()

# 2. Sugar Level vs BMI
plt.figure(figsize=(6,4))
colors = df['Disease_Status'].map({'Yes':'red','No':'green'})
plt.scatter(df['Sugar_Level'], df['BMI'], c=colors)
plt.title("Sugar Level vs BMI")
plt.xlabel("Sugar Level")
plt.ylabel("BMI")
plt.show()

# 3. Blood Pressure vs Sugar Level
plt.figure(figsize=(6,4))
plt.scatter(df['BP'], df['Sugar_Level'], c=colors)
plt.title("Blood Pressure vs Sugar Level")
plt.xlabel("Blood Pressure")
plt.ylabel("Sugar Level")
plt.show()

# 4. Disease Status Bar Chart
df['Disease_Status'].value_counts().plot(kind='bar')
plt.title("Disease Status")
plt.xlabel("Disease")
plt.ylabel("Number of Patients")
plt.show()

# Correlation Matrix
print("\nCorrelation Matrix")
print(df[['Age','Sugar_Level','BP','BMI']].corr())

# -----------------------------
# EDA Conclusion
# -----------------------------
print("\nEDA Conclusion")
print("1. Patients with high sugar levels are more likely to have disease.")
print("2. Higher blood pressure is commonly seen in diseased patients.")
print("3. Higher BMI is associated with disease risk.")
print("4. Older patients generally show higher health risk.")
print("5. Sugar level, BP, and BMI are important factors affecting disease status.")
```
