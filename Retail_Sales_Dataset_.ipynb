```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create Dataset
data = {
    'Product_ID': ['PR01','PR02','PR03','PR04','PR05','PR06','PR07','PR08','PR09','PR10',
                   'PR11','PR12','PR13','PR14','PR15','PR16','PR17','PR18','PR19','PR20'],
    'Category': ['Grocery','Snacks','Dairy','Fruits','Vegetables','Grocery','Snacks','Dairy','Fruits','Vegetables',
                 'Grocery','Snacks','Dairy','Fruits','Vegetables','Grocery','Snacks','Dairy','Fruits','Vegetables'],
    'Price': [50,30,45,60,40,80,25,55,70,35,90,20,65,75,45,100,35,50,80,30],
    'Quantity_Sold': [20,35,18,25,30,15,40,16,22,28,12,45,14,20,26,10,32,19,18,35],
    'Discount': [5,3,2,5,4,6,3,2,5,4,6,2,3,5,4,7,3,2,6,3],
    'Revenue': [1000,1050,810,1500,1200,1200,1000,880,1540,980,1080,900,910,1500,1170,1000,1120,950,1440,1050]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display Dataset
print("Retail Sales Dataset")
print(df)

# Dataset Information
print("\nDataset Information")
print(df.info())

# Summary Statistics
print("\nSummary Statistics")
print(df.describe())

# Missing Values
print("\nMissing Values")
print(df.isnull().sum())

# Category Count
print("\nProducts by Category")
print(df['Category'].value_counts())

# Revenue by Category
print("\nRevenue by Category")
print(df.groupby('Category')['Revenue'].sum())

# -----------------------------------
# Data Visualization
# -----------------------------------

# 1. Revenue by Category
plt.figure(figsize=(7,5))
df.groupby('Category')['Revenue'].sum().plot(kind='bar')
plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Total Revenue")
plt.show()

# 2. Price vs Revenue
plt.figure(figsize=(6,4))
plt.scatter(df['Price'], df['Revenue'])
plt.title("Price vs Revenue")
plt.xlabel("Price")
plt.ylabel("Revenue")
plt.show()

# 3. Quantity Sold vs Revenue
plt.figure(figsize=(6,4))
plt.scatter(df['Quantity_Sold'], df['Revenue'])
plt.title("Quantity Sold vs Revenue")
plt.xlabel("Quantity Sold")
plt.ylabel("Revenue")
plt.show()

# 4. Discount Distribution
plt.figure(figsize=(6,4))
plt.hist(df['Discount'], bins=6)
plt.title("Discount Distribution")
plt.xlabel("Discount (%)")
plt.ylabel("Number of Products")
plt.show()

# Correlation Matrix
print("\nCorrelation Matrix")
print(df[['Price','Quantity_Sold','Discount','Revenue']].corr())

# -----------------------------------
# Highest Selling Product
# -----------------------------------
highest_quantity = df.loc[df['Quantity_Sold'].idxmax()]
print("\nHighest Selling Product")
print(highest_quantity)

# Highest Revenue Product
highest_revenue = df.loc[df['Revenue'].idxmax()]
print("\nHighest Revenue Product")
print(highest_revenue)

# -----------------------------------
# EDA Conclusion
# -----------------------------------
print("\nEDA Conclusion")
print("1. Fruits category generates the highest revenue.")
print("2. Products with higher quantity sold generally produce higher revenue.")
print("3. Higher-priced products do not always generate the highest revenue.")
print("4. Moderate discounts help increase product sales.")
print("5. Quantity sold and price are the main factors affecting revenue.")
```
